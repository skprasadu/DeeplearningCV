```python
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

img = load_img('dog.jpeg')  
x = img_to_array(img)  # creating a Numpy array with shape (3, 150, 150)\n",
x = x.reshape((1,) + x.shape)  # converting to a Numpy array with shape (1, 3, 150, 150)\n",

i = 0
for batch in datagen.flow(x, save_to_dir='output', save_prefix='dog', save_format='jpeg'):
    i += 1
    if i > 35:
        break 
```
